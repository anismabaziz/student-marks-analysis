```python
import pandas as pd
import re
```

```python
df = pd.read_csv("../data/students_data.csv")
df.head()
```

```python
# adding section column
i = 0

def extract_number(text):
    match = re.search(r'[\d\u0660-\u0669]+', text)
    return int(match.group()) if match else None

for index, row in df.iterrows():
  student_num = int(extract_number(row["Name"]))
  if student_num == 1:
    i += 1
  df.at[index, "Section"] = i
```

```python
# cutting the name
def extract_name(text):
    if text:
        text = re.sub(r'^\d+\s*-\s*', '', text.strip())
    return text

df["Name"] = df["Name"].apply(extract_name)

df.head()
```
