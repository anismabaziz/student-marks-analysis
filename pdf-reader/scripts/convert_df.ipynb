```python
import pandas as pd
import re
```

```python
file_name = "Deliberation_2024_2025_L2_S3"
df = pd.read_csv(f"../data/{file_name}.csv")
df.head()
```

```python
# cutting the name
def extract_name(text):
    if text:
        text = re.sub(r'^\d+\s*-\s*', '', text.strip())
    return text

df["Name"] = df["Name"].apply(extract_name)

df.head()
```

```python
df.to_csv(f"../data/{file_name}_postprocess.csv", index=False)
```
